## Exhaustive Selection  (SFS)

```python
import pandas as pd
from mlxtend.feature_selection import ExhaustiveFeatureSelector
```

### Import dataset

```python
df = pd.read_csv("Finaltest.csv")
```

```python
df.head()
```

### Drop columns

```python
#df=df.drop(columns='id_student')
df=df.drop(columns='code_module')
df=df.drop(columns='code_presentation')
df=df.drop(columns='date_unregistration')
df=df.drop(columns='date_registration')
```

```python
#check row and columns
df.shape
```

```python
#Check missing value
df.isnull().sum()
```

```python
#check if any missing value
df[df.isnull().any(axis=1)]
```

```python
X = df.iloc[:,0:22] # all columns/features except target 
y = df.iloc[:,-1]  #target
```

```python
X
```

```python
y
```

```python
X.head()
```

### Train and Fit dataset

```python
from sklearn.model_selection import train_test_split
#training 0.3,
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

### Exhaustive Selection 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X_train.shape
```

```python
efs = ExhaustiveFeatureSelector(GaussianNB(),
                                min_features=1,
                                max_features=10,#how model is fitted
                                scoring= "f1",   #performance matric
                                cv= 5).fit(X_train, y_train)   #cross validation
```

```python
efs.best_feature_names_
```

```python
efs.best_score_
```

```python
pd.DataFrame.from_dict(efs.get_metric_dict()).T
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import matplotlib.pyplot as plt

fig1 = plot_sfs(efs.get_metric_dict(), kind='std_dev')

plt.title('Sequential Forward Selection')
plt.grid()
plt.show()
```
