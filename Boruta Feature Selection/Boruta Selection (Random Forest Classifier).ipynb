## Boruta Selection (SFS)

```python
 import pandas as pd
 import numpy as np
 from sklearn.ensemble import RandomForestClassifier
 from boruta import BorutaPy
 from sklearn.model_selection import train_test_split
 from sklearn.metrics import f1_score
```

### Import dataset

```python
df = pd.read_csv("Finaltest.csv")
```

```python
df.head()
```

### Drop columns

```python
#df=df.drop(columns='id_student')
df=df.drop(columns='code_module')
df=df.drop(columns='code_presentation')
df=df.drop(columns='date_unregistration')
df=df.drop(columns='date_registration')
```

```python
#check row and columns
df.shape
```

```python
#Check missing value
df.isnull().sum()
```

```python
#check if any missing value
df[df.isnull().any(axis=1)]
```

```python
X = df.iloc[:,0:22] # all columns/features except target 
y = df.iloc[:,-1]  #target
```

```python
X
```

```python
y
```

```python
X.head()
```

### Train and Fit dataset

```python
from sklearn.model_selection import train_test_split
#training 0.3,
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

### Sequential Forward Selection

```python
X_train.shape
```

```python
 rf_all_features = RandomForestClassifier(random_state=1, n_estimators=1000, max_depth=5)
 rf_all_features.fit(X_train, y_train) 
```

```python
f1_score(y_test, rf_all_features.predict(X_test))
```

```python
 rfc = RandomForestClassifier(random_state=1, n_estimators=1000, max_depth=5)
 boruta_selector = BorutaPy(rfc, n_estimators='auto', verbose=2, random_state=1)
 boruta_selector.fit(np.array(X_train), np.array(y_train)) 
```

```python
 print("Ranking: ",boruta_selector.ranking_)          
 print("No. of significant features: ", boruta_selector.n_features_) 
```

```python
selected_rf_features = pd.DataFrame({'Feature':list(X_train.columns),
                                       'Ranking':boruta_selector.ranking_})
selected_rf_features.sort_values(by='Ranking') 
```

```python
X_important_train = boruta_selector.transform(np.array(X_train))
X_important_test = boruta_selector.transform(np.array(X_test))
```

```python
rf_boruta = RandomForestClassifier(random_state=1, n_estimators=1000, max_depth=5)
rf_boruta.fit(X_important_train, y_train)
```

```python
f1_score(y_test, rf_boruta.predict(X_important_test))
```
