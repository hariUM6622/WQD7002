```python
one_hot_encoded_data = pd.get_dummies(df_log, columns = ['source','event'])
```

```python
result = one_hot_encoded_data.groupby('enrollment_id')['source_browser','source_server','event_access','event_discussion','event_navigate','event_page_close','event_problem','event_video','event_wiki'].aggregate('sum')
```
